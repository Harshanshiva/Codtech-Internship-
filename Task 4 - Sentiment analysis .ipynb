```python
# Task 4: Sentiment Analysis using TextBlob

# ðŸ“¦ Step 1: Import required libraries
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
import os

# ðŸ“‚ Step 2: Load the dataset (update path if needed for Databricks)
df = pd.read_csv("product_reviews_sample.csv")  # Replace with uploaded path if needed

# ðŸ§  Step 3: Define a function to get sentiment polarity
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# ðŸŽ¯ Step 4: Apply sentiment analysis to the reviews
df['sentiment_score'] = df['review_text'].apply(get_sentiment)

# ðŸ·ï¸ Step 5: Label each review as Positive, Neutral, or Negative
df['sentiment_label'] = df['sentiment_score'].apply(
    lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral')
)

# ðŸ“Š Step 6: Display results
print("\nSentiment Analysis Results:")
print(df[['review_text', 'sentiment_score', 'sentiment_label']])

# ðŸ“‰ Step 7: Visualize sentiment distribution
sentiment_counts = df['sentiment_label'].value_counts()
sentiment_counts.plot(kind='bar', title="Sentiment Distribution", color=['green', 'gray', 'red'])
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()

# ðŸ’¾ Step 8: (Optional) Save results to CSV for submission
# Save to the current directory instead of a non-existent path
df.to_csv("sentiment_analysis_results.csv", index=False)
# Alternatively, you can create the directory first if you have permissions
# import os
# os.makedirs("/dbfs/FileStore/tables", exist_ok=True)
# df.to_csv("/dbfs/FileStore/tables/sentiment_analysis_results.csv", index=False)

# âœ… Task Complete
```
